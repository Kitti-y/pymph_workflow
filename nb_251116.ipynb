```python
import pandas as pd

df1_read = pd.read_csv('coordinates_spreadsheets\251118_coor_grid6_3_6_6_nonzero.csv')

client = mph.start()
model = client.load('251117_template_ellipsoid_GAGs.mph')

modeljava = model.java

## create for loops for all configs

#radius_list = [5,6,7,8]

for index, row in df1_read.head(2).iterrows():
    r_coord = row['r']
    xdeg_coord = row['x_deg']
    tdeg_coord = row['t_deg']
    pdeg_coord = row['p_deg']

    r_index = int(row['nr'])
    x_index = int(row['nx'])
    t_index = int(row['nt'])
    p_index = int(row['np'])

# for radius in radius_list:

    # change parameter r_coord
    modeljava.param().set('rposition', str(r_coord))
    modeljava.param().set('angle_xi', str(xdeg_coord))
    modeljava.param().set('angle_theta', str(tdeg_coord))
    modeljava.param().set('angle_phi', str(pdeg_coord))
    
    modeljava.component('comp1').geom('geom1').run()

    modeljava.study('std3').run()

    #modeljava.save(f'run_test/251117_template_{r_index}{x_index}{t_index}{p_index}.mphbin')
    model.save(f'run_test/251117_template_{r_index}{x_index}{t_index}{p_index}.mph')

    #modeljava.component('comp1').geom('geom1').exportFinal(f'export_test/test_export_{r_index}{x_index}{t_index}{p_index}.mphbin')
    # modeljava.component('comp1')
model.clear()
```
